```python
#차대차 사건중 죽거나 다친사람이 많은 발생지 시도를 알아보고 시각화
#1. 차대차 사건중 죽거나 다친사람(사상자)이 많은 발생지 시도
#2. 시각화
```

```python
# 차대차 사건만 출력
acc =  pn[pn["사고유형_대분류"] == '차대차']
acc
# acc["사고유형_대분류"]
# acc['발생지시도'].value_counts()
```

```python
acc['발생지시도'].unique()
```

```python
acc['사상자수'].value_counts()
```

```python
#그룹화
#group by() -> 이상태로는 아무것도 안나온다
#groupby().연산자() -> sum()이나 avg()이나 이런걸 넣어줘야 결과가 나온다.
#그룹화 하면 그룹화 기준이 되는 컬럼은 인덱스가 된다
#그룹화를 두번이상 할 수 있는데 처음에 쓴걸 기준으로 그룹화가 되고 그다음이 그룹화 된다.
#불러올땐 변수명['대그룹','소그룹']
a = acc[['사상자수', '발생지시도']].groupby("발생지시도").sum() 
type(a)
```

```python
acc_g = acc[['사상자수', '발생지시도']].groupby("발생지시도").sum()
x = acc_g.index
y = acc_g['사상자수']
plt.figure(figsize = (10,5)) #가로 세로 크기 지정
plt.title("차vs차 교통사고의 사상자 수")
plt.ylabel('사상자 수', rotation = 0)
plt.xlabel('지역')
plt.bar(x,y)
```
