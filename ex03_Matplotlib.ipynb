```python
import matplotlib.pyplot as plt
```

# 선그래프

```python
#plot 함수 - 일직선으로 되어있는 Lineplot그리기
# 설정된 수에 점을 찍고 점을 이어준것!
y = [2,4,6]
plt.plot(y) # 그래프 그리기(y축을 기준으로 그래지고 x축 값이 설정이 안되어 있으면 자동으로 짝을 맞추어 생성된다.)
plt.show() # 그래프 출력
```

```python
#x축 설정해서 그래프 그리기
#축을 두개다 설정해줄때는 짝이 맞아야 한다!
x = [1,2,3]
y = [2,4,6]
# 설정된 수에 점을 찍고 점을 이어준것!
plt.plot(x,y)
plt.show()
```

```python
#line.style
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y, linestyle = "--")
```

```python
x = [2,4,6,8]
y = [10,13,16,20]
y2 = [10,20,16,13]
plt.plot(x,y)
plt.plot(x,y2, ls=":")
plt.show()
```

```python
#maker
plt.plot(x,y, marker = "1")
```

```python
#line width -> lw라는 약어로 사용 가능
#라인의 두께 지정
plt.plot(x,y,linewidth="5")
```

```python
#color
#따로 지정하지 않으면 마커의 색깔도 같이 바뀐다.
plt.plot(x,y, color = "red", marker = "8")
```

```python
#makerfacecolor -> 약어는 mfc
plt.plot(x,y, color = "red", marker = "8", markerfacecolor = "blue")
```

```python
#실습
x = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, ls = "--", color = "green", marker = "8")
plt.show
```

```python
plt.plot(x,y, ls = ":", color = "red", marker = "s", mfc = "blue")
```

```python
#실습
import numpy as np

x = np.array(x)
y = [1,4,5,8,9,5,3]
plt.plot(x,y, ls='--', marker ="o", color = "blue", mfc = "orange", lw = "5", ms = "15", mec = "g", mew = "5")
```
